# Importing experiment data

```python
%matplotlib inline
from pyfitit import *

project = LoadProject('FeterpyProject_old.py', expType = 'highspin-excited')
```

# Importing XANES calculated for support points

```python
sample = readSample('IHS_500')
```

# Construct the iverse estimator

```python
convolutionParams = {'Gamma_hole':6.1305, 'Ecent':27.771, 'Elarg':1, 'Efermi':10.075, 'Gamma_max':13.238, 'shift':7112.6, 'norm':None}
inverseEstimator = constructInverseEstimator('Extra Trees', project, convolutionParams, CVcount=2, folderToSaveCVresult='inverseEstimator_CVresult')
```

# Compare different machine learning methods

```python
inverseEstimator.compareDifferentMethods(sample, folderToSaveResult = 'result')
```

# Train the estimator

```python
inverseEstimator.fit(sample)
```

# Start automatic fitting

```python
inverseMethod.findGlobalL2NormMinimum(1, inverseEstimator, folderToSaveResult = 'inverseMethodResults')
```
